# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,1,11)
y = x**2
```

```python
# Functional
plt.plot(x,y)
plt.xlabel('X-Label')
plt.ylabel('Y=Label')
plt.title('Title')
# plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r-')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oreinted Method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X-Label')
axes.set_ylabel('Y-Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

# axes.plot(x,y)
# for i in axes:
#     i.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title("First")

axes[1].plot(y,x)
axes[1].set_title("Second")

plt.tight_layout()
```

# Figure size and DPI

```python
fig = plt.figure(figsize= (8,2))

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows= 2, ncols= 1, figsize=(8,5))

axes[0].plot(x,y)
axes[0].set_title('Title')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

axes[1].plot(y,x)
axes[1].set_title('Title')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')

plt.tight_layout()
```

```python
fig.savefig('My_picture.png', dpi=200)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2, label='X Square')
axes.plot(x,x**3, label='X Cubed')
axes.set_title('Title')
axes.set_xlabel('X')
axes.set_ylabel('Y')

axes.legend(loc=0)
# axes.legend(loc=(0.1,0.1))
```

# Plt Appearrence

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y, color='#FF8C00', linewidth='2', alpha=1, linestyle='--', marker='o', markersize=10, 
         markerfacecolor='red', markeredgewidth=3, markeredgecolor='black')   # RGB Hex Code

axes.set_xlim([0,1])
axes.set_ylim([0,1])
```
