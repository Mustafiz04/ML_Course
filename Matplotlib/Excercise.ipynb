```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)

axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,0.2,0.2])

axes1.plot(x,y, color='r')
axes2.plot(x,y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,0.4,0.4])

axes1.plot(x,z)
axes1.set_xlabel('X')
axes1.set_ylabel('Z')

axes2.plot(x,y)
axes2.set_xlabel("X")
axes2.set_ylabel("Y")
axes2.set_xlim([20,22])
axes2.set_ylim([30,50])
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(12,2))

axes[0].plot(x,y, linewidth=3,color='b')
axes[1].plot(x,y,linestyle='--', linewidth=3, color='r')
```
