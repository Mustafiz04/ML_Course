# Matrix Plot

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flight = sns.load_dataset('flights')
tips.head()
```

```python
flight.head()
```

```python
tc = tips.corr()
tc
```

```python
sns.heatmap(tc, annot=True)
```

```python
fb = flight.pivot_table(index='month',columns='year',values='passengers')
fb
```

```python
sns.heatmap(fb,linecolor='white',linewidths=1)
```

```python
sns.clustermap(fb,standard_scale=1)
```
